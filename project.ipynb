```python
!pip install ucimlrepo

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, RepeatedStratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from ucimlrepo import fetch_ucirepo
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from sklearn.metrics import accuracy_score, f1_score, matthews_corrcoef, recall_score, precision_score, confusion_matrix
from prettytable import PrettyTable

breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)

X = breast_cancer_wisconsin_diagnostic.data.features
y = breast_cancer_wisconsin_diagnostic.data.targets

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y.values.ravel())

# Standardize the input features
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Define a function to create and train the model
def create_and_train_model(optimizer_name, activation_function, dropout_rate, X_train, y_train, X_test, y_test):
    # Define the architecture of the neural network
    model = Sequential()

    # Input layer with dynamic input dimension
    input_dim = X_train.shape[1]
    model.add(Dense(units=30, input_dim=input_dim))

    # Hidden layer 1
    model.add(Dense(units=30))
    model.add(Activation(activation_function))
    model.add(Dropout(dropout_rate))

    # Hidden layer 2
    model.add(Dense(units=15))
    model.add(Activation(activation_function))
    model.add(Dropout(dropout_rate))

    # Hidden layer 3
    model.add(Dense(units=30))
    model.add(Activation(activation_function))

    # Output layer
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(optimizer=optimizer_name, loss='binary_crossentropy', metrics=['accuracy'])

    model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)

    y_pred_prob = model.predict(X_test)
    y_pred = (y_pred_prob > 0.5).astype(int)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    cm = confusion_matrix(y_test, y_pred)

    tn, fp, fn, tp = cm.ravel()

    sensitivity = recall_score(y_test, y_pred)
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0


    precision = precision_score(y_test, y_pred, zero_division=1)

    recall = recall_score(y_test, y_pred)

    return accuracy, f1, mcc, sensitivity, specificity, precision, recall

optimizers = ["Adadelta", "Adagrad", "Adam", "Adamax", "Nadam", "RMSprop", "SGD"]
activation_functions = ["swish"]

table = PrettyTable()
table.field_names = ["DNN", "Optimizer", "Accuracy", "F1-Score", "MCC", "Sensitivity", "Specificity", "Precision", "Recall"]

n_splits = 10
n_repeats = 1
kf = RepeatedStratifiedKFold(n_splits=n_splits, n_repeats=n_repeats, random_state=42)

for activation_function in activation_functions:
    for optimizer in optimizers:
        if activation_function == "linear" and optimizer == "SGD":
            continue

        accuracies, f1_scores, mcc_scores, sensitivities, specificities, precisions, recalls = [], [], [], [], [], [], []

        for train_index, test_index in kf.split(X_standardized, y):
            X_train, X_test = X_standardized[train_index], X_standardized[test_index]
            y_train, y_test = y[train_index], y[test_index]

            accuracy, f1, mcc, sensitivity, specificity, precision, recall = create_and_train_model(optimizer, activation_function, dropout_rate=0.3, X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test)

            accuracies.append(accuracy)
            f1_scores.append(f1)
            mcc_scores.append(mcc)
            sensitivities.append(sensitivity)
            specificities.append(specificity)
            precisions.append(precision)
            recalls.append(recall)

        mean_accuracy = np.mean(accuracies) * 100
        mean_f1 = np.mean(f1_scores)
        mean_mcc = np.mean(mcc_scores)
        mean_sensitivity = np.mean(sensitivities)
        mean_specificity = np.mean(specificities)
        mean_precision = np.mean(precisions)
        mean_recall = np.mean(recalls)

        table.add_row([
            f"{activation_function.capitalize()}DNN",
            optimizer,
            mean_accuracy,
            mean_f1,
            mean_mcc,
            mean_sensitivity,
            mean_specificity,
            mean_precision,
            mean_recall
        ])

print(table)
```

Extention: Simple two-hidden-layer architecture with 128 and 64 units.
Utilizes the Rectified Linear Unit (ReLU) activation function in hidden layers.
Applies dropout with a rate of 0.3 after each hidden layer to prevent overfitting. Uses a binary sigmoid activation function in the output layer.

```python
!pip install ucimlrepo

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, RepeatedStratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from ucimlrepo import fetch_ucirepo
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, f1_score, matthews_corrcoef, recall_score, precision_score, confusion_matrix
from tensorflow.keras.callbacks import EarlyStopping
from prettytable import PrettyTable

breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)

X = breast_cancer_wisconsin_diagnostic.data.features
y = breast_cancer_wisconsin_diagnostic.data.targets

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y.values.ravel())

scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

def create_and_train_model(optimizer_name, activation_function, dropout_rate, X_train, y_train, X_test, y_test):
    model = Sequential()

    model.add(Dense(units=128, input_dim=X_train.shape[1], activation=activation_function))
    model.add(Dropout(dropout_rate))

    model.add(Dense(units=64, activation=activation_function))
    model.add(Dropout(dropout_rate))

    model.add(Dense(units=1, activation='sigmoid'))

    model.compile(optimizer=optimizer_name, loss='binary_crossentropy', metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), callbacks=[early_stopping], verbose=0)

    y_pred_prob = model.predict(X_test)
    y_pred = (y_pred_prob > 0.5).astype(int)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()

    sensitivity = recall_score(y_test, y_pred)
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0
    precision = precision_score(y_test, y_pred, zero_division=1)
    recall = recall_score(y_test, y_pred)

    return accuracy, f1, mcc, sensitivity, specificity, precision, recall

optimizers = ["Adam", "Adagrad", "Nadam", "RMSprop"]

table = PrettyTable()
table.field_names = ["DNN", "Optimizer", "Accuracy", "F1-Score", "MCC", "Sensitivity", "Specificity", "Precision", "Recall"]

n_splits = 10
n_repeats = 1
kf = RepeatedStratifiedKFold(n_splits=n_splits, n_repeats=n_repeats, random_state=42)

for optimizer in optimizers:
    accuracies, f1_scores, mcc_scores, sensitivities, specificities, precisions, recalls = [], [], [], [], [], [], []

    for train_index, test_index in kf.split(X_standardized, y):
        X_train, X_test = X_standardized[train_index], X_standardized[test_index]
        y_train, y_test = y[train_index], y[test_index]

        accuracy, f1, mcc, sensitivity, specificity, precision, recall = create_and_train_model(
            optimizer, activation_function='relu', dropout_rate=0.3, X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test
        )

        accuracies.append(accuracy)
        f1_scores.append(f1)
        mcc_scores.append(mcc)
        sensitivities.append(sensitivity)
        specificities.append(specificity)
        precisions.append(precision)
        recalls.append(recall)

    mean_accuracy = np.mean(accuracies) * 100
    mean_f1 = np.mean(f1_scores)
    mean_mcc = np.mean(mcc_scores)
    mean_sensitivity = np.mean(sensitivities)
    mean_specificity = np.mean(specificities)
    mean_precision = np.mean(precisions)
    mean_recall = np.mean(recalls)

    table.add_row([
        "DNN",
        optimizer,
        mean_accuracy,
        mean_f1,
        mean_mcc,
        mean_sensitivity,
        mean_specificity,
        mean_precision,
        mean_recall
    ])

# Print the table
print(table)
```
